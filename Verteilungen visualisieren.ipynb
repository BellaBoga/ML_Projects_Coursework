```python
import numpy as np

levels = np.random.normal(300, 10, 10000)
print(levels)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.hist(levels, bins = 100)
plt.xlim(100, 500)
plt.show()
```
